```python
stocks_and_etfs=pd.read_csv("Downloads\\Spring_Invitational_2024_data\\Datathon Data\\all_stocks_and_etfs.csv")
stocks_and_etfs.loc[:,'Date-Time']=pd.to_datetime(stocks_and_etfs.loc[:,'Date-Time'])

stocks_and_etfs=stocks_and_etfs.sort_values('Date-Time')
stocks_and_etfs
```

```python
stocks={ticker: stocks_and_etfs[stocks_and_etfs['Ticker_Symbol']==ticker] for ticker in (set(stocks_and_etfs['Ticker_Symbol']))}
stocks['HAL'].plot(x='Date-Time',y='Close')
```

```python
for ticker in set(stocks_and_etfs['Ticker_Symbol']):
    stocks[ticker].loc[:,'daily_return']=stocks[ticker].loc[:,'Close'].pct_change()
    stocks[ticker]=stocks[ticker].dropna()
```

```python
stocks['HAL']
```

```python
quarterly_returns=pd.DataFrame(columns=list(set(commods['Commodity']))+list(set(stocks_and_etfs['Ticker_Symbol'])),
                               index=['Q_'+str(quarter)+'_'+str(year)  for year in range(2005,2024) for quarter in range(1,5)])

for ticker in set(stocks_and_etfs['Ticker_Symbol']):
    for year in range(2005,2024):
        for quarter in range(1,5):
            data=stocks[ticker]
            data=data[data['Date-Time'].between(pd.Timestamp(year=year,month=3*quarter-2,day=1),pd.Timestamp(year=year,month=3*quarter,day=30))]
            if(data.shape[0]<1):
                continue
            start_price=data.sort_values(by='Date-Time',ascending=True).iloc[0,4]
            end_price=data.sort_values(by='Date-Time',ascending=True).iloc[-1,4]
            quarterly_returns.loc['Q_'+str(quarter)+'_'+str(year),ticker]=(end_price-start_price)/start_price
            
for ticker in set(commods['Commodity']):
    for year in range(2005,2024):
        for quarter in range(1,5):
            data=commods[commods['Commodity']==ticker]
            data=data[data['Date_Time'].between(pd.Timestamp(year=year,month=3*quarter-2,day=1),pd.Timestamp(year=year,month=3*quarter,day=30))]
            if(data.shape[0]<1):
                continue
            start_price=data.sort_values(by='Date_Time',ascending=True).iloc[0,1]
            end_price=data.sort_values(by='Date_Time',ascending=True).iloc[-1,1]
            quarterly_returns.loc['Q_'+str(quarter)+'_'+str(year),ticker]=(end_price-start_price)/start_price
            
quarterly_returns
```

```python
quarterly_returns.to_csv('quarterly_returns.csv')
quarterly_returns.corr().to_csv('Quarterly_return_correlations.csv')
```

```python
quarterly_vol=pd.DataFrame(columns=list(set(commods['Commodity']))+list(set(stocks_and_etfs['Ticker_Symbol'])),
                               index=['Q_'+str(quarter)+'_'+str(year)  for year in range(2005,2024) for quarter in range(1,5)])

for ticker in set(stocks_and_etfs['Ticker_Symbol']):
    for year in range(2005,2024):
        for quarter in range(1,5):
            data=stocks[ticker]
            data=data[data['Date-Time'].between(pd.Timestamp(year=year,month=3*quarter-2,day=1),pd.Timestamp(year=year,month=3*quarter,day=30))]
            if(data.shape[0]<1):
                continue
            quarterly_vol.loc['Q_'+str(quarter)+'_'+str(year),ticker]=data['daily_return'].std()
            
for ticker in set(commods['Commodity']):
    for year in range(2005,2024):
        for quarter in range(1,5):
            data=commods[commods['Commodity']==ticker]
            data=data[data['Date_Time'].between(pd.Timestamp(year=year,month=3*quarter-2,day=1),pd.Timestamp(year=year,month=3*quarter,day=30))]
            if(data.shape[0]<1):
                continue

            quarterly_vol.loc['Q_'+str(quarter)+'_'+str(year),ticker]=data['Daily Return'].std()
            
quarterly_vol
```

```python
quarterly_vol.to_csv('quarterly_vol.csv')
quarterly_vol.corr().to_csv('Quarterly_vol_correlations.csv')
```
